```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset['Phone']=dataset['Phone'].str.replace("-","")
dataset['Phone']=dataset['Phone'].astype(int)
```

```python
dataset.head()
```

```python
X=dataset.columns[dataset.dtypes=="O"]
X.values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in X.values:
    dataset[i]=le.fit_transform(dataset[i].astype(str))
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.columns
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
new=ss.fit_transform(dataset.drop('Churn?',axis=1).values)
```

```python
data=pd.DataFrame(data=new,columns=['State', 'Account Length', 'Area Code', 'Phone', "'Int'l Plan",
       'VMail Plan', 'VMail Message', 'Day Mins', 'Day Calls', 'Day Charge',
       'Eve Mins', 'Eve Calls', 'Eve Charge', 'Night Mins', 'Night Calls',
       'Night Charge', 'Intl Mins', 'Intl Calls', 'Intl Charge',
       'CustServ Calls'])
data.head()
```

```python
cor=data.corrwith(dataset.iloc[:,-1])
cor
```

```python
X=data[['State', 'Account Length', 'Area Code', "'Int'l Plan",
       'Day Mins', 'Day Calls', 'Day Charge', 'Eve Mins', 'Eve Calls',
       'Eve Charge', 'Night Mins', 'Night Calls', 'Night Charge', 'Intl Mins',
       'Intl Charge', 'CustServ Calls']]
```

```python
Y=dataset["Churn?"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=2000,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
knn.score(X_train,Y_train)
```

```python
pred=knn.predict(X_test)
knn.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(confusion_matrix(Y_test,pred))
```

```python
print(classification_report(Y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
clf =LogisticRegression()
fitted_model = clf.fit(X_train, Y_train)
```

```python
clf.score(X_train, Y_train)
```

```python
prediction=clf.predict(X_test)
```

```python
clf.score(X_test,Y_test)
```

```python
tc=DecisionTreeClassifier(max_depth=4)
tc.fit(X_train,Y_train)
```

```python
tc.score(X_train,Y_train)
```

```python
pred1=clf.predict(X_test)
```
