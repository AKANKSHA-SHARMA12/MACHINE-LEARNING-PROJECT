# Creating dataframe of rating and no.of rating


```python
rating=pd.DataFrame(data.groupby('title')['rating'].mean())
rating.head()
```

```python
rating['No of rating']=pd.DataFrame(data.groupby('title')['rating'].count())
rating.head()
```

# Recommending Movie

```python
matrix=data.pivot_table(index='user_id',columns='title',values='rating')
matrix.head()
```

```python
user_rating=matrix["Star Wars (1977)"]
user_rating.head()
```

```python
correlation_matrix=matrix.corrwith(matrix["Star Wars (1977)"])
```

```python
correlation_matrix=correlation_matrix.dropna()


```

```python
recommend_movie=pd.DataFrame(correlation_matrix,columns=["Correlation"])
recommend_movie.sort_values('Correlation',ascending=False).head()
```

```python
recommend_movie=recommend_movie.join(rating["No of rating"])
```

```python
recommend_movie.sort_values("Correlation" , ascending=False).head()
```

```python
recommend_movie[recommend_movie['No of rating']>200].sort_values("Correlation" , ascending=False).head()
```

# Same analysis for Liar Liar movie
